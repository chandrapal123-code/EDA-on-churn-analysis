```python
import statistics

def find_anomalies(df):
    anomalies = []
    # Mean aur standard deviation nikalna
    random_df_std = statistics.stdev(df)
    random_df_mean = statistics.mean(df)

    # 3 sigma rule (mean ± 3*std)
    anomaly_cut_off = random_df_std * 3
    lower_limit = random_df_mean - anomaly_cut_off
    upper_limit = random_df_mean + anomaly_cut_off

    # Outliers detect karna
    for outlier in df:
        if outlier > upper_limit or outlier < lower_limit:
            anomalies.append(outlier)
    
    return anomalies
```
